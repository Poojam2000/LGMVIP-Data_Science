## Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
iris["Species"] = le.fit_transform(iris['Species'])
iris.head()
```

## Model Training

```python
x = iris.drop(['Species'],1)
x.head()
```

```python
y = iris['Species']
y.head()
```

```python
#test-train split
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

### Model 1: Logistics Regression Model

```python
from sklearn.linear_model import LogisticRegression
model_lr = LogisticRegression()
```

```python
model_lr.fit(x_train,y_train)
```

```python
print("Accuracy of the Logistic Regression Model :",model_lr.score(x_test,y_test)*100)
```

### Model 2 : Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
model_dtc = DecisionTreeClassifier()
```

```python
model_dtc.fit(x_train,y_train)
```

```python
print("Accuracy of the Decision Tree Classifier Model :", model_dtc.score(x_test, y_test)*100)
```

### Model 3 : KNN K-Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
model_knn = KNeighborsClassifier()
```

```python
model_knn.fit(x_train, y_train)
```

```python
print("Accuracy of the KNN K-Nearest Neighbors Model ", model_knn.score(x_test, y_test)*100)
```
